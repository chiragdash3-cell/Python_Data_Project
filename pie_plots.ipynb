```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
from datasets import load_dataset
dataset=load_dataset('lukebarousse/data_jobs')
df= dataset["train"].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list )
```

```python
df['job_no_degree_mention'].value_counts().plot(kind="pie", startangle=90, autopct="%1.1f%%")
```

```python
fig,ax = plt.subplots(1,3)


dict_column= {
    'job_work_from_home':'work_from_home',
    'job_no_degree_mention':'no_degree_req',
    'job_health_insurance':'health_insurance_provided'
}

for i,(column,title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct="%.2f%%",labels=[False,True])
    ax[i].set_title(title)

plt.show()    
```
